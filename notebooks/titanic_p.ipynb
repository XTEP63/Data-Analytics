```python
import pandas as pd 
import sweetviz as sv

df_train = pd.read_csv("../data/raw/titanic/train.csv")
df_test = pd.read_csv("../data/raw/titanic/test.csv")
```

```python
df_train.head(3)
```

```python
def clean(data):
    data = data.drop(["Ticket", "Cabin", "Name"], axis = 1)
    
    cols = ["SibSp", "Parch", "Fare", "Age"]
    
    for col in cols:
        data[col].fillna(data[col].median(), inplace= True)
    
    return data 
```

```python
data = clean(df_train)
test = clean(df_test)
```

```python
y_mean = round(data["Survived"].mean())
test["Survived"] = y_mean

for index, row in test.iterrows():
    if row["SibSp"] == 2 or row["SibSp"] == 1:
        test.at[index, "Survived"] = int(1)
        

test
```

```python
test["Survived"] = test["Survived"].astype(int)
test["PassengerId"] = test["PassengerId"].astype(int)
```
