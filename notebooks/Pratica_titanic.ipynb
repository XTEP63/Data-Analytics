```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = "../data/raw/tested.csv"
df = pd.read_csv(path)
df = df.drop(columns=['Cabin'], inplace=False)
df.columns
```

Podemos observar la cantidad de personas que hay por edad

```python
sns.histplot(df, x = "Age", color= "skyblue")
```

Vemos un histograma en la que nos pmuestra los daos separados por clase en los distintos colores de las barras y pero que tambien separa estos por si sobrevivieron o no 

```python
sns.histplot(df, x = "Survived", hue= "Pclass", multiple= "dodge")
```

Vemos que hay hombres en cada rango edad sin a exepcion de 4 rangos espesificos 

```python
sns.histplot(df, x = "Age", hue= "Sex")
```

Vemos que en genereal la mayoria de pasajerosno contaban con hermanos

```python
sns.histplot(df, x = "Embarked", hue="SibSp", palette="deep")
```

Vemos que la mayoria de los pasajeros no pagraon precios altos

```python
sns.histplot(df["Fare"], bins= 20, kde = True)
```

La mayoria de padres no tenia hijos

```python
sns.histplot(df["Parch"])
```

La mayoria no tenia hermanos o tenian 1 

```python
sns.histplot(df["SibSp"], binwidth = 1, color = "red")
```

```python
df_D = df[df["Survived"] == 0]
df_D
```

```python
df_V = df[df["Survived"] == 1]
df_V
```

Vemos un gran decremento de las personas que sobrevivieron segun entre mayores sean, pero al mismo tiempo vemos que la mayoria de muertes son de la gente que esta entre las edades medias 

```python
sns.histplot(data=df_V, x="Age", bins=10, alpha=0.7, color='blue')
sns.histplot(data=df_D, x="Age", bins=10, alpha=0.3, color='red')
```
